```python
f.seek(0)
a=f.readlines()
```

```python
x=[]
y=[]
for i in a:
    if(i[0]=='+'):
        t=2
        tt=""
        while(i[t]!=' '):
            tt+=i[t]
            t+=1
        
        if(i[t+24]=="1"):
            x.append(tt)
    if(i[0]=='-'):
        t=2
        tt=""
        while(i[t]!=' '):
            tt+=i[t]
            t+=1
        if(i[t+24]=="1"):
            y.append(tt)
        
f.close()
```

```python
print(x)
```

```python
print(y)
```

```python
res=[]
for i in range(len(x)):
    print(float(y[i])-float(x[i]))
    tt=(x[i])+" "+str(round(((float(y[i])-float(x[i]))),5))
    res.append(tt)
```

```python
print(res)
```

```python
f=open("q.txt","w")
for i in res:
    f.write(i+"\n")
f.close()
```
